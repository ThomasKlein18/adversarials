```python
# RUN THIS FILE LAST, AFTER SIMPLE AND CNN

import numpy as np
import tensorflow as tf
import pickle
import matplotlib.pyplot as plt
```

```python
# paths

global_path = "/Users/thomasklein/Uni/5. Semester/TensorFlow/final task/"
datagraph_path = global_path+"graph_data/"
```

```python
file = open(datagraph_path+"regularSimple.pkl", 'rb')
regularSimple = pickle.load(file)
file.close()

file = open(datagraph_path+"adversarialSimple.pkl", 'rb')
adversarialSimple = pickle.load(file)
file.close()

file = open(datagraph_path+"regularCNN.pkl", 'rb')
regularCNN = pickle.load(file)
file.close()

file = open(datagraph_path+"adversarialCNN.pkl", 'rb')
adversarialCNN = pickle.load(file)
file.close()

file = open(datagraph_path+"xval.pkl", 'rb')
xvals = pickle.load(file)
file.close()
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(xvals,regularSimple, 'r', label="regular simple")
plt.plot(xvals, adversarialSimple, 'b', label="adversarial simple")
plt.plot(xvals, regularCNN, 'g', label="regular CNN")
plt.plot(xvals, adversarialCNN, 'y', label="adversarial CNN")
plt.ylabel("Correct Classifications in %")
plt.xlabel("Epsilon")
plt.title("Network Accuracy")
plt.legend()
plt.show()
fig.savefig("Accuracy_vs_Epsilon.jpg")
```
